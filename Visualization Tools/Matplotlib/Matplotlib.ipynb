```python
subjects = ['Maths', 'English', 'Science', 'Social Science', 'Computer']
marks_1 = [89,65,97,78,97]
marks_2 = [23,45,76,33,35]
```

```python
plt.figure(figsize = (8,8))

plt.bar(subjects, marks_1)
plt.bar(subjects, marks_2)

plt.xlabel('Subjects')
plt.ylabel('Marks')

plt.show()
```

```python
subjects_len = np.arange(len(subjects))
width = 0.4
```

```python
plt.figure(figsize = (8,8))

plt.bar(subjects_len, marks_1, width = width)
plt.bar(subjects_len + width, marks_2, width = width)

plt.xlabel('Subjects')
plt.ylabel('Marks')

plt.show()
```

```python
plt.figure(figsize = (8,8))

plt.bar(subjects_len, marks_1, width = width, color = colors)
plt.bar(subjects_len + width, marks_2, width = width, color = colors, alpha = 0.5)

plt.xlabel('Subjects')
plt.ylabel('Marks')

plt.show()
```

```python
marks_50_students = np.random.randint(0,100,(50))
marks_50_students
```

```python
plt.hist(marks_50_students)
plt.show()
```

```python
bins = np.arange(0,100,5)

plt.figure(figsize = (12,6))

plt.hist(marks_50_students, bins = bins)

plt.xticks(np.arange(0,100,5))
plt.show()
```

```python
bins = np.arange(0,100,5)

plt.figure(figsize = (12,6))

plt.hist(marks_50_students, bins = bins, color = 'orange')

plt.xticks(np.arange(0,100,5))
plt.show()
```

```python
bins = np.arange(0,100,5)

plt.figure(figsize = (12,6))

plt.hist(marks_50_students, bins = bins, color = 'orange', orientation = 'horizontal', rwidth = 0.4)

plt.yticks(np.arange(0,100,5))
plt.show()
```

```python
bins = np.arange(0,100,5)

plt.figure(figsize = (6,6))

plt.hist(marks_50_students, bins = bins, color = 'orange', histtype = 'step')

plt.xticks(np.arange(0,100,5))
plt.show()
```

```python
classes = ['Physics', 'Chemistry', 'Maths', 'SST', 'Hindi']
marks = [23,54,78,67,89]
```

```python
plt.pie(marks, labels = classes)
plt.show()
```

```python
colors = ['red', 'green', 'orange', 'blue', 'purple']

plt.pie(marks, labels = classes, colors = colors)
plt.show()
```

```python
plt.pie(marks, labels = classes, colors = colors, autopct = '%0.1f%%')
plt.show()
```

```python
explode_values = [0.1, 0.2, 0, 0, 0]

plt.pie(marks, labels = classes, colors = colors, autopct = '%0.1f%%', explode = explode_values)
plt.show()
```

```python
plt.pie(marks, labels = classes, colors = colors, autopct = '%0.1f%%', explode = explode_values, shadow = True)
plt.show()
```

```python
plt.pie(marks, labels = classes, colors = colors, autopct = '%0.1f%%', explode = explode_values, radius = 1.6)
plt.show()
```

```python
textprops = {'fontsize':14, 'color':'w'}

plt.pie(marks, labels = classes, colors = colors, autopct = '%0.1f%%', explode = explode_values, radius = 1.6, textprops = textprops)
plt.show()
```

```python
wedgeprops = {'linewidth': 3, 'edgecolor':'black'}

plt.pie(marks, labels = classes, colors = colors, autopct = '%0.1f%%', explode = explode_values, radius = 1.6, textprops = textprops, wedgeprops = wedgeprops)
plt.show()
```

```python
plt.figure(figsize = (6,6))

plt.pie(marks, labels = classes, colors = colors, autopct = '%0.1f%%', explode = explode_values, radius = 1.6, textprops = textprops)

plt.legend()

plt.show()
```
